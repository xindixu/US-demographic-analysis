### Acs_2018 Data Analysis

##### To analyze and see a relationship between college and undergraduate enrollment in public vs private and the predominant class in the area/neighborhood

##### Analysis the relationship between the predominant income class and its housing value class for each area/neighborhood

##### Analysis the relationship between average income, social classes and predominant industry

```python
%%bigquery
select sum(Rich) as Total_Rich, sum(Upper_Middle) as Total_Upper_Middle, sum(Middle) as Total_Middle, 
avg(Average) as Average, Predominant_Industry_First, Predominant_Industry_Second, Predominant_Industry_Third
from `acs_2018_modeled.Income_Beam_DF` i join `acs_2018_modeled.Labor_In_Industry_Beam_DF` l on i.ZCTA5 = l.ZCTA5 
where Predominant_Class = "Rich" or Predominant_Class = "Upper_Middle" or Predominant_Class = "Middle"
group by Predominant_Industry_First,Predominant_Industry_Second, Predominant_Industry_Third
having Predominant_Industry_First <> "" and Predominant_Industry_Second <> ""  and Predominant_Industry_Third <> ""
order by Average desc
limit 20
```

##### Analysis the relationship between working weeks per year, working hours per week and average income

```python
%%bigquery 
select Work_50_to_52_weeks, w.No_working as no_working_yearly, Work_35h_or_more, h.No_working as no_working_weekly, Average
from `acs_2018_modeled.Working_Hours_Per_Week_Beam_DF` h 
left join `acs_2018_modeled.Working_Weeks_Per_Year_Beam_DF` w on h.ZCTA5 = w.ZCTA5 
left join `acs_2018_modeled.Income_Beam_DF` i on h.ZCTA5 = i.ZCTA5
where Average <> 0
order by Average desc
limit 20
```

##### Analysis total number of working from home and total number of working full week in different predominant industry

```python
%%bigquery 
select sum(Work_From_Home) as Total_Work_From_Home, sum(Work_35h_or_more) as Total_Full_Work_Week, Predominant_Industry_First, Predominant_Industry_Second, Predominant_Industry_Third
from `acs_2018_modeled.Working_Hours_Per_Week_Beam_DF` h 
join  `acs_2018_modeled.Commute_Beam_DF` c on c.ZCTA5 = h.ZCTA5 
join  `acs_2018_modeled.Labor_In_Industry_Beam_DF` i on i.ZCTA5 = h.ZCTA5 
group by Predominant_Industry_First, Predominant_Industry_Second, Predominant_Industry_Third
having Total_Full_Work_Week > 0
order by Total_Full_Work_Week desc, Total_Work_From_Home desc
limit 20
```

## Acs_2018 Data Analysis with Subqueries

##### Analysis of area/neighborhood that has the number of college/undergrad and graduate student higher than the average number of students in each category.

```python
%%bigquery
select ZCTA5 as ZCTA5_higher_than_average_schoolenrollment, College_Undergrad, Grad_HigherEdu
from acs_2018_modeled.General_School_Enrollment_Beam_DF
where College_Undergrad >
(select avg(College_Undergrad) as average_College_Undergrad
from acs_2018_modeled.General_School_Enrollment_Beam_DF)
and Grad_HigherEdu >
(select avg(Grad_HigherEdu) as average_Grad_HigherEd
from acs_2018_modeled.General_School_Enrollment_Beam_DF)
order by ZCTA5
limit 20
```

##### Analysis of which industry is predominant at the area/neighborhood that has number of graduate enrollment higher than the average enrollments

```python
%%bigquery
select Predominant_Industry_First, count(Predominant_Industry_First) as Area_with_Industry
from
(select l.ZCTA5, Predominant_Industry_First
from acs_2018_modeled.Labor_In_Industry_Beam_DF l join acs_2018_modeled.General_School_Enrollment_Beam_DF s on l.ZCTA5 = s.ZCTA5
where Grad_HigherEdu > 
(select avg(Grad_HigherEdu) as average_Grad_HigherEd
from acs_2018_modeled.General_School_Enrollment_Beam_DF))
where Predominant_Industry_First is not null
group by Predominant_Industry_First
order by Area_with_Industry desc
```

##### Analysis of which industry is predominant at the area/neighborhood has `Rich` or `Upper_Middle` as its predominant socialeconomic classes and housing value are high

```python
%%bigquery
select Predominant_Industry_First, count(*) as Number_of_Area
from `acs_2018_modeled.Labor_In_Industry_Beam_DF` 
where ZCTA5 in
(select i.ZCTA5 
 from `acs_2018_modeled.Income_Beam_DF` i
 join `acs_2018_modeled.Housing_Value_Beam_DF` h
 on i.ZCTA5 = h.ZCTA5 
 where (Predominant_Class = 'Rich' or Predominant_Class = 'Upper_Middle') and High_Housing_Value > 0)
group by Predominant_Industry_First
order by Number_of_Area desc
```

#### Analysis of housing value in the area/neighborhood has `Rich` as its predominant socialeconomic classes

```python
%%bigquery
select High_Housing_Value, Middle_Housing_Value, Low_Housing_Value
from `acs_2018_modeled.Housing_Value_Beam_DF`  
where ZCTA5 in
(select ZCTA5 from `acs_2018_modeled.Income_Beam_DF` 
where Predominant_Class = 'Rich')
order by High_Housing_Value desc
```
