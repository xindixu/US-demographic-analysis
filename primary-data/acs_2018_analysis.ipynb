### Acs_2018 Data Analysis

##### Analysis of a relationship between college and undergraduate enrollment in public vs private and the predominant class of Income in the area/neighborhood

```python
%%bigquery
select Predominant_Class as Predominant_Income_Class, 
sum(College_Undergrad_Public) as Total_College_Undergrad_Public, 
sum(College_Undergrad_Private) as Total_College_Undergrad_Private, 
sum(Grad_HigherEdu_Public) as Total_Grad_HigherEdu_Public, 
sum(Grad_HigherEdu_Private) as Total_Grad_HigherEdu_Private
from acs_2018_modeled.Income_Beam_DF i join acs_2018_modeled.PrivatePublic_School_Enrollment_Beam_DF s
on i.ZCTA5 = s.ZCTA5
group by Predominant_Class 
order by Total_College_Undergrad_Private desc
limit 20
```

##### Analysis of housing ages and housing value

```python
%%bigquery
select b.*, High_Housing_Value
from acs_2018_modeled.Built_Year_Beam_DF b join acs_2018_modeled.Housing_Value_Beam_DF h
on b.ZCTA5 = h.ZCTA5
where High_Housing_Value <> 0 and Middle_Housing_Value <> 0 and Low_Housing_Value <> 0
order by High_Housing_Value desc
limit 20
```

##### Analysis of the number of house built in 2041 or later that has a high housing value in each area/neighborhood

```python
%%bigquery
select High_Housing_Value, count(Built_2014_or_later) as num_of_highvalue_built2014_or_later
from acs_2018_modeled.Built_Year_Beam_DF b join acs_2018_modeled.Housing_Value_Beam_DF h
on b.ZCTA5 = h.ZCTA5
where High_Housing_Value <> 0 and Middle_Housing_Value <> 0 and Low_Housing_Value <> 0
group by High_Housing_Value having High_Housing_Value > 50
order by num_of_highvalue_built2014_or_later desc
limit 20
```

Analysis the relationship between average income, social classes and predominant industry

```python
%%bigquery
select sum(Rich) as Total_Rich, sum(Upper_Middle) as Total_Upper_Middle, sum(Middle) as Total_Middle, 
avg(Average) as Average, Predominant_Industry_First, Predominant_Industry_Second, Predominant_Industry_Third
from `acs_2018_modeled.Income_Beam_DF` i join `acs_2018_modeled.Labor_In_Industry_Beam_DF` l on i.ZCTA5 = l.ZCTA5 
where Predominant_Class = "Rich" or Predominant_Class = "Upper_Middle" or Predominant_Class = "Middle"
group by Predominant_Industry_First,Predominant_Industry_Second, Predominant_Industry_Third
having Predominant_Industry_First <> "" and Predominant_Industry_Second <> ""  and Predominant_Industry_Third <> ""
order by Average desc
limit 20
```

Analysis the relationship between working weeks per year, working hours per week and average income

```python
%%bigquery 
select Work_50_to_52_weeks, w.No_working as no_working_yearly, Work_35h_or_more, h.No_working as no_working_weekly, Average
from `acs_2018_modeled.Working_Hours_Per_Week_Beam_DF` h 
left join `acs_2018_modeled.Working_Weeks_Per_Year_Beam_DF` w on h.ZCTA5 = w.ZCTA5 
left join `acs_2018_modeled.Income_Beam_DF` i on h.ZCTA5 = i.ZCTA5
where Average <> 0
order by Average desc
limit 20
```

Analysis total number of working from home and total number of working full week in different predominant industry

```python
%%bigquery 
select sum(Work_From_Home) as Total_Work_From_Home, sum(Work_35h_or_more) as Total_Full_Work_Week, Predominant_Industry_First, Predominant_Industry_Second, Predominant_Industry_Third
from `acs_2018_modeled.Working_Hours_Per_Week_Beam_DF` h 
join  `acs_2018_modeled.Commute_Beam_DF` c on c.ZCTA5 = h.ZCTA5 
join  `acs_2018_modeled.Labor_In_Industry_Beam_DF` i on i.ZCTA5 = h.ZCTA5 
group by Predominant_Industry_First, Predominant_Industry_Second, Predominant_Industry_Third
having Total_Full_Work_Week > 0
order by Total_Full_Work_Week desc, Total_Work_From_Home desc
limit 20
```

## Create view

```python
%%bigquery 
create or replace view acs_2018_modeled.v_Industry_With_Higher_Income as
select i.ZCTA5, sum(Rich) as Total_Rich, sum(Upper_Middle) as Total_Upper_Middle, sum(Middle) as Total_Middle, 
avg(Average) as Average, Predominant_Industry_First, Predominant_Industry_Second, Predominant_Industry_Third
from `sashimi-266523.acs_2018_modeled.Income_Beam_DF` i 
join `sashimi-266523.acs_2018_modeled.Labor_In_Industry_Beam_DF` l on i.ZCTA5 = l.ZCTA5 
where Predominant_Class = "Rich" or Predominant_Class = "Upper_Middle" or Predominant_Class = "Middle"
group by i.ZCTA5, Predominant_Industry_First,Predominant_Industry_Second, Predominant_Industry_Third
having Predominant_Industry_First <> "" and Predominant_Industry_Second <> ""  and Predominant_Industry_Third <> ""
order by Average desc
```
