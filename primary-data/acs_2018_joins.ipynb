# ACS 2018 Join Queries

% of population in labor force, % household income 50K-74K, and house value 200K-299K, filtered out 100% and null values, ordered by labor force, income, then housing value (full join)

```python
%%bigquery
select lfs.NAME, lfs.DP03_0002PE as Labor_Force, i.DP03_0057PE as Income_50K_74K, hv. DP04_0085PE as House_value_200K_299K
from `acs_2018_modeled.Labor_Force_Stats` as lfs 
full outer join `acs_2018_modeled.Income` as i on lfs.NAME = i.NAME
full outer join acs_2018_modeled.Housing_Value as hv
on i.NAME = hv.NAME
order by lfs.NAME
limit 20
```

% houses are Mobile home and RV or Boats, % of population in the labor_force, % of income 15K-24K, % of incom 100K-149K (full join)

```python
%%bigquery
select i.NAME, hu.DP04_0014PE as Mobile_Home, hu.DP04_0015PE as RV_Boat, lfs.DP03_0002PE as Labor_Force, 
i.DP03_0054PE as Income_15K_24K, i.DP03_0059PE as Income_100K_149K
from `acs_2018_modeled.Income` as i 
full join acs_2018_modeled.Housing_Unit as hu on i.NAME = hu.NAME
full join acs_2018_modeled.Labor_Force_Stats lfs on hu.NAME = lfs.NAME
order by i.NAME
limit 20
```

People in "white-collar" industry and their housing values, ordered by % of housing value 1M_Or_More (full join)

```python
%%bigquery
select lii.NAME, DP04_0088PE as Housing_Value_1M_Or_More, DP03_0039E as Information, 
DP03_0040E as Finance_Real_Estate, DP03_0041E as Prof_Sci_Man, 
DP03_0042E as Educational_Health_Care, DP03_0043E as Art_Entertainment 
from acs_2018_modeled.Labor_In_Industry as lii 
full join `acs_2018_modeled.Housing_Value` as hu on lii.NAME = hu.NAME
order by DP04_0088PE desc
limit 10
```

% of population has high incom 100K-149K and 150K-199K, college and above education, and "white-collar" industry, order by % has income 150K-199K (full join)

```python
%%bigquery
select i.NAME, DP03_0059PE as Income_100K_149K, DP03_0060PE as Income_150K_199K, S1401_C01_008E as College_Undergrad, 
S1401_C01_009E as Grad_HigherEd, DP03_0039E as Information, DP03_0040E as Finance_Real_Estate, 
DP03_0041E as Professional_Scientific_Management, DP03_0042E as Educational_Health_Care, 
DP03_0043E as Art_Entertainment, DP03_0045E as Public_Administration
from acs_2018_modeled.Income as i 
full join `acs_2018_modeled.General_School_Enrollment` as gse on i.NAME = gse.NAME
full join acs_2018_modeled.Labor_In_Industry as lii on gse.NAME = lii.NAME
order by i.NAME
limit 20
```

The query shows whether there is a connection between the percentage of people in different range of income and the distribution in public and private college and undegraduate enrollment

```python
%%bigquery
select ppse.NAME, DP03_0052PE as Income_less_10K, DP03_0053PE as Income_10K_14K, DP03_0054PE as Income_15K_24K,
DP03_0055PE as Income_25K_34K, DP03_0056PE as Income_35K_49K, DP03_0057PE as Income_50K_74K,
DP03_0058PE as Income_75K_99K, DP03_0059PE as Income_100K_149K, DP03_0060PE as Income_150K_199K,
DP03_0061PE as Income_200K_more, S1401_C03_008E as College_Undergrad_Public,
S1401_C05_008E as College_Undergrad_Private
from acs_2018_modeled.PrivatePublic_School_Enrollment ppse join acs_2018_modeled.Income i on ppse.NAME = i.NAME
order by i.NAME
limit 20
```

The query shows the connection of daily hours of working and the different range of income

```python
%%bigquery
select whpw.NAME, DP03_0052PE as Income_less_10K, DP03_0053PE as Income_10K_14K, DP03_0054PE as Income_15K_24K,
DP03_0055PE as Income_25K_34K, DP03_0056PE as Income_35K_49K, DP03_0057PE as Income_50K_74K,
DP03_0058PE as Income_75K_99K, DP03_0059PE as Income_100K_149K, DP03_0060PE as Income_150K_199K,
DP03_0061PE as Income_200K_more, S2303_C01_009E as Work_35h_or_more, S2303_C01_016E as Work_16h_to_34h,
S2303_C01_023E as Work_1h_to_15h, S2303_C01_030E as No_working
from acs_2018_modeled.Working_Hours_Per_Week whpw join acs_2018_modeled.Income i on whpw.NAME = i.NAME
order by whpw.NAME
limit 20
```

The relationship between neighborhood whose family earns high income (>75K), has higher housing value (>400K) and the working weeks per year (full join)

```python
%%bigquery
select i.NAME, DP03_0058PE as Income_75K_99K, DP03_0059PE as Income_100K_149K, DP03_0060PE as Income_150K_199K,
DP03_0061PE as Income_200K_more, DP04_0086PE as Housing_Value_400K_499K, DP04_0087PE as Housing_Value_500K_999K, 
DP04_0088PE as Housing_Value_1M_Or_More, S2303_C01_002E as Work_50_to_52_weeks, S2303_C01_003E as Work_48_to_49_weeks,
S2303_C01_004E as Work_40_to_47_weeks, S2303_C01_005E as Work_27_to_39_weeks, S2303_C01_006E as Work_14_to_26_weeks,
S2303_C01_007E as Work_1_to_13_weeks, S2303_C01_008E as No_working_Yearly
from acs_2018_modeled.Income i 
full join acs_2018_modeled.Housing_Value hu on i.NAME = hu.NAME
full join acs_2018_modeled.Working_Weeks_Per_Year wwpy on hu.NAME = wwpy.NAME
order by i.NAME
limit 20
```
