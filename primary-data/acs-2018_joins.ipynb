% of population in labor force, % household income 50K-74K, and house value 200K-299K, filtered out 100% and null values, ordered by labor force, income, then housing value

```python
%%bigquery
select e.NAME, DP03_0002PE as Labor_Force, DP03_0057PE as Income_50K_74K, DP04_0085PE as V_200K_299K
from `acs_2018_staging.economics` as e 
join `acs_2018_staging.housing` as h 
on e.NAME = h.NAME
where DP03_0002PE != '-' and DP03_0057PE != '-' and DP04_0085PE != '-' and DP03_0002PE != '100' and DP03_0057PE != '100' and DP04_0085PE != '100' 
order by cast(DP03_0002PE as FLOAT64) desc, cast(DP03_0057PE as FLOAT64) desc, cast(DP04_0085PE as FLOAT64) desc
limit 20
```
