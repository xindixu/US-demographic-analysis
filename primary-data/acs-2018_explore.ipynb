## Economics 

```python
%%bigquery
select NAME, DP03_0002PE as Labor_Force, DP03_0014E as Children_In_HouseHold from acs_2018_staging.economics
where DP03_0014E is not null
order by DP03_0014E desc
limit 10
```

```python
%%bigquery
select NAME, DP03_0033E as Argiculture, DP03_0034E as Construction, DP03_0035E as Manufacturing, 
DP03_0036E as Wholesale, DP03_0037E as Retail, DP03_0038E as Transportation, DP03_0039E as Information, 
DP03_0040E as Finance_Real_Estate, DP03_0041E as Professional_Scientific_Management, 
DP03_0042E as Educational_Health_Care, DP03_0043E as Art_Entertainment, DP03_0045E as Public_Administration
from acs_2018_staging.economics
where DP03_0003E is not null
order by DP03_0003E desc
limit 10
```

## Housing

```python
%%bigquery
select NAME, DP04_0001E as Total_Housing_Units, DP04_0002PE as Occupied_Housing_Unites, DP04_0003PE as Vacant_Housing_Unites
from acs_2018_staging.housing
where DP04_0001E is not null
order by DP04_0002PE desc
limit 10
```

% of housing that's occupied and % of housing vacant by ZCTA

```python
%%bigquery
select NAME, DP04_0006E as Total_Housing_Units, DP04_0007PE as Detached_Unit_1, DP04_0008PE as Attached_Unit_1, DP04_0009PE as Units_2, DP04_0010PE as Unites_3_4, 
DP04_0011PE as Unites_5_9, DP04_0012PE as Unites_10_19, DP04_0013PE as Unites_20_More, DP04_0014PE as Mobile_Home, DP04_0015PE as RV_Boat
from acs_2018_staging.housing
where safe_cast(DP04_0015PE as FLOAT64) > 0
order by safe_cast(DP04_0015PE as FLOAT64) desc
limit 20
```

% of housing in different types, ordered by % of RV, Boats

```python
%%bigquery
select NAME, DP04_0017PE as Built_2014_Or_Later, DP04_0018PE as Built_2010_2013, DP04_0019PE as Built_2000_2009, DP04_0020PE as Built_1990_1999, DP04_0021PE as Built_1980_1989, 
DP04_0022PE as Built_1970_1979, DP04_0023PE as Built_1960_1969, DP04_0024PE as Built_1950_1959, DP04_0025PE as Built_1940_1949, DP04_0026PE as Built_1939_Or_After
from acs_2018_staging.housing
where safe_cast(DP04_0017PE as FLOAT64) > 0
order by safe_cast(DP04_0017PE as FLOAT64) desc
limit 10
```

% of houses built within different periods of time, order by houses built later than 2014

```python
%%bigquery
select NAME, DP04_0081PE as Less_Than_50k, DP04_0082PE as V_50k_99K, DP04_0083PE as V_100K_149K, DP04_0084PE as V_150K_199K, DP04_0085PE as V_200K_299K, 
DP04_0086PE as V_400K_499K, DP04_0087PE as V_500K_999K, DP04_0088PE as V_1M_Or_More
from acs_2018_staging.housing
where safe_cast(DP04_0088PE as FLOAT64) > 0
order by safe_cast(DP04_0088PE as FLOAT64) desc
limit 100
```
