## school enrollment

```python
# explore about the numbers of school enrollment in categories for each ZCTA area
```

```python
%%bigquery
select NAME, S1401_C01_004E as Kindergarden, S1401_C01_005E as Grade1_to_4, S1401_C01_006E as Grade5_to_8, S1401_C01_007E as Grade9_to_12, S1401_C01_008E as College_Undergrad, S1401_C01_009E as Grad_HigherEdu from acs_2018_staging.school_enrollment
where S1401_C01_004E is not null
order by NAME
limit 20
```

```python
# explore about the school enrollment in college or higher for public school and private school
```

```python
%%bigquery
select NAME, S1401_C03_008E as College_Undergrad_Public, S1401_C05_008E as College_Undergrad_Private, S1401_C03_009E as Grad_HigherEdu_Public, S1401_C05_009E as Grad_HigherEdu_Private from acs_2018_staging.school_enrollment
where S1401_C03_008E is not null
order by NAME
limit 20
```

```python
# explore about female and male ratio in school enrollment for college and higher education 
# in public school and private school
```

```python
%%bigquery
select NAME, S1401_C03_011E as Male_Public, S1401_C05_011E as Female_Public, S1401_C03_012E as Male_Private, S1401_C05_012E as Female_Private from acs_2018_staging.school_enrollment
where S1401_C03_011E is not null
order by NAME
limit 20
```

## work

```python
### summary the work hours of each ZCTA area
```

```python
%%bigquery
select NAME, S2303_C01_031E as Mean_hours, S2303_C01_032E as Median_age_working, S2303_C01_033E as Full_time_year_round from acs_2018_staging.work
where S2303_C01_031E is not null
order by NAME
limit 20
```

```python
# explore about the number of working hours for each ZCTA area
```

```python
%%bigquery
select NAME, S2303_C01_009E as Work_35h_or_more, S2303_C01_016E as Work_16h_to_34h, S2303_C01_023E as Work_1h_to_15h, S2303_C01_030E as No_working from acs_2018_staging.work
where S2303_C01_016E is not null
order by NAME
limit 20
```

```python
# the ratio between male and female who did not work
```

```python
%%bigquery
select NAME, S2303_C03_030E as Male_no_working, S2303_C05_030E as Female_no_working from acs_2018_staging.work
where S2303_C03_030E is not null
order by NAME
limit 20
```
